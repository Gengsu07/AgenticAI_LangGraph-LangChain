### Load model

```python
model = ChatGoogleGenerativeAI(model='models/gemini-flash-latest')
model_groq = ChatGroq(model='llama-3.3-70b-versatile')
```

### Messages

```python
from langchain_core.messages import SystemMessage, HumanMessage

messages = [
    SystemMessage('You are expert explainer, please expain given concept fundamentally step by step in easy to digest way'),
    HumanMessage('How aircraft can fly?')
]
```

### Inference

```python
result = model.invoke(messages)
```

```python
result1 = model_groq.invoke(messages)
```

```python
result1.content
```

### Parser ke string

```python

from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(result)
```

```python
parser.invoke(result1)
```

```python
from langchain_core.output_parsers import MarkdownListOutputParser
md = MarkdownListOutputParser()
md.invoke(result1)
```

```python
md.invoke(result)
```

### Pakai Template

```python
from langchain_core.prompts import ChatPromptTemplate

Instruct_template = "You are cultural young Indonesian, please translate teks into bahasa gaul di {kota}"

prompt = ChatPromptTemplate.from_messages(
    [("system",Instruct_template),("user","{text}")]
)

hasilprompt = prompt.invoke({"kota":"Jakarta Selatan","text":"Hello, please introduce my name is Aksa. I live in Jaksel"})
hasilprompt.to_messages()
```

### Langchain Expression Language (LCEL)

```python
chain = model|parser
result = chain.invoke('how to master langchain and langgraph')
result
```

```python
chainwithtemplate = prompt|model_groq|parser
result1 = chainwithtemplate.invoke(({"kota":"Jakarta Selatan","text":"Hello, please introduce my name is Aksa. I live in Jaksel"}))
```
