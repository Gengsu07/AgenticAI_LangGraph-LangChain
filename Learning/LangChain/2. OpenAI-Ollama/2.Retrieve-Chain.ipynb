#### Embedding dengan Google

List embedding model
- text-embedding-004
- text-multilingual-embedding-002

Kapan pilih text-multilingual-embedding-002?
Pilihlah model ini jika:

Konteks Bahasa Indonesia: Model ini dilatih khusus untuk memahami nuansa semantik dalam banyak bahasa, termasuk Bahasa Indonesia. Hasil pencarian (retrieval) biasanya jauh lebih akurat untuk query non-Inggris.
Cross-Language Search: Jika Anda ingin user bertanya dalam Bahasa Indonesia tetapi mencari dokumen yang mungkin ada dalam Bahasa Inggris (atau sebaliknya).
Efisiensi Biaya: Jauh lebih murah untuk pemrosesan dokumen dalam skala besar.
Kapan pilih text-embedding-004?
Pilihlah model ini jika:

Full English: Aplikasi Anda sepenuhnya dalam Bahasa Inggris.
Code Search: Anda membangun tool untuk mencari potongan kode pemrograman.
Dokumen Panjang: Mendukung input token sedikit lebih banyak.

```python
from google import genai
import os

# 1. Inisialisasi Client
client = genai.Client()

# 2. List semua model yang tersedia
# print("--- Daftar Semua Model ---")
# for model in client.models.list():
#     print(f"Name: {model.name}")
#     print(f"Supported Actions: {model.supported_actions}")
#     print("-" * 20)

# 3. Filter khusus untuk model Embedding
print("\n--- Model yang Mendukung Embedding ---")
for model in client.models.list():
    if 'embedContent' in model.supported_actions:
        print(f"Embedding Model Found: {model.name}")
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embeddings_google = GoogleGenerativeAIEmbeddings(model="models/gemini-embedding-001", task_type="retrieval_document")
```

Perbedaan task_type (Kenapa ini penting?)
Google menyarankan penggunaan parameter task_type untuk hasil yang lebih akurat:

- retrieval_document: Gunakan saat Anda melakukan embedding pada dokumen yang akan disimpan di vector database.
- retrieval_query: Gunakan saat Anda melakukan embedding pada pertanyaan user (query) untuk mencari dokumen yang relevan.
- semantic_similarity: Gunakan jika Anda hanya ingin membandingkan kesamaan teks biasa.
Tips Pro: Jika Anda menggunakan LangChain secara standar, model akan mencoba menyesuaikan secara otomatis, tetapi menentukan task_type secara eksplisit seringkali memberikan hasil pencarian yang lebih tajam.

### Retrieve-Chain & Document Chain

https://docs.langchain.com/oss/python/integrations/chat

```python
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_ollama import ChatOllama

# llm_openai = ChatOpenAI(model="gpt-4o", temperature=0)
llm_google_genai = ChatGoogleGenerativeAI(model="gemini-3-flash-preview", temperature=0)
llm_ollama = ChatOllama(model="gemma3:4b", temperature=0)
```

https://reference.langchain.com/python/langchain_core/prompts/#langchain_core.prompts.chat.ChatPromptTemplate

create_stuff_documents_chain is likely referring to the StuffDocumentsChain from LangChain's legacy chains, now in @langchain/classic/chains.

This chain concatenates multiple documents into a single prompt for the LLM (the "stuff" approach).

In LangChain v1, legacy chains like this have moved to the @langchain/classic package. Install it with pip install langchain-classic.

```python
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template(
    """
    Tolong jawab pertanyaan berikut hanya berdasarkan konteks yang diberikan:
    <context>
    {context}
    </context>
    """
)

document_chain = create_stuff_documents_chain(llm_google_genai, prompt)
document_chain_ollama = create_stuff_documents_chain(llm_ollama, prompt)
document_chain
```


Input (documents + question)
    ↓

[1] RunnableAssign: format_docs
    → Memformat dokumen menjadi context string
    ↓

[2] ChatPromptTemplate
    → Membuat prompt: "Jawab berdasarkan context: {context}"
    ↓

[3] ChatGoogleGenerativeAI
    → LLM memproses prompt dan menghasilkan jawaban
    ↓

[4] StrOutputParser
    → Mengekstrak jawaban sebagai string
    ↓
    
Output (string answer)


### Cari konteks secara dinamis dari Retriever

```python
retriever_ollama = vecdb_ollama.as_retriever()
retriever_google = vecdb_google.as_retriever()
```

```python
from langchain_classic.chains import create_retrieval_chain

ollama_retrievalchain = create_retrieval_chain(
    retriever_ollama,
    document_chain_ollama
)

google_retrievalchain = create_retrieval_chain(
    retriever_google,
    document_chain
)
```

```python
response_google = google_retrievalchain.invoke({'input':'What is database?'})
response_google
```
